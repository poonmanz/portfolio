```
def extract_feat(df):
  feat = df.copy()
  
  # ‡∏•‡∏≠‡∏á‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏°‡∏≤‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏ó‡∏µ‡πà‡∏°‡∏µ catergory ‡πÄ‡∏ó‡πà‡∏≤‡∏Å‡∏±‡∏ô‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á train ‡πÅ‡∏•‡∏∞ test ‡πÑ‡∏î‡πâ‡πÅ‡∏Å‡πà brand, transmission ‡πÅ‡∏•‡∏∞ car_type
  dummy_feat = pd.get_dummies(df[['transmission','car_type','brand']])
  feat = pd.concat([feat, dummy_feat], axis=1)

  # dummies_df = pd.get_dummies(feat['brand'])
  # feat = pd.concat([feat,dummies_df],axis=1)
  feat = feat.drop(['brand'],axis=1)

  label = feat['price']
  
  drop_columns = ['location','price','desc','ad_id','timestamp','model','year','fuel','color','transmission','car_type']
  feat = feat.drop(drop_columns, axis=1)

  
  return feat, label
```

```
train_feat, train_label = extract_feat(train_df)
test_feat, test_label = extract_feat(test_df)
```

# Train Model

```
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error,r2_score
```

```
def train_model(feat, label, lr, es):
  model = GradientBoostingRegressor(random_state=0, learning_rate=lr, n_estimators=es)
  model.fit(feat, label)
  return model
```

```
from sklearn import tree
def train_tree_model(feat, label, md):
  model = tree.DecisionTreeRegressor(max_depth=md)
  model.fit(feat,label)
  return model
```

```
learning_rate = np.array(range(1,10,1))/10.0
print(learning_rate)
for lr in learning_rate:
  model = train_model(train_feat, train_label, lr, es=100 )# error ‡πÅ‡∏£‡∏Å ‡∏ô‡πà‡∏≤‡∏à‡∏∞‡∏•‡∏∑‡∏°‡πÉ‡∏™‡πà‡∏Ñ‡πà‡∏≤ es ‡∏Ñ‡∏£‡∏±‡∏ö
  predict = model.predict(test_feat)
  mse = mean_squared_error(predict, test_label)
  r2 = r2_score(predict, test_label)
  print("lr =",lr,"mse =", mse,"r2score =" ,r2 )
```

```
max_depth = range(1,20)
for md in max_depth:
  model = train_tree_model(train_feat, train_label, md)
  predict = model.predict(test_feat)
  mse = mean_squared_error(predict, test_label)
  print(md, mse)
```
